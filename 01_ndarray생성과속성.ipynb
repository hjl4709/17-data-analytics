```python
# 배열 속성

arr1 = np.array([1, 2, 3, 4])
print(arr1.shape)       # (4, ) -> 1차원, 4행

arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print(arr2.shape)       # (2, 3) -> 2차원, 2행 3열

arr3 = np.array([
    [[1, 2],
     [3, 4]],
     [[5, 6],
      [7, 8]]
])
print(arr3.shape)       # (2, 2 ,2) -> 3차원 / 텐서
```

```python
# ndim -> 차원 출력
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
```

```python
# size - 배열의 총 요소 수를 알려준다.

arr1 = np.array([1, 2, 3, 4])
print(arr1.size)

arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print(arr2.size)

arr3 = np.array([
    [[1, 2],
     [3, 4]],
     [[5, 6],
      [7, 8]]
])
print(arr3.size)
```
