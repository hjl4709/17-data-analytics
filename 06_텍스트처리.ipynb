```python
import numpy as np
import pandas as pd

ser = pd.Series(["Apple_사과", "Banana_바나나", "Cherry_체리", np.nan],
                index=["First ", " Second", " Third", "Fourth"])
print(ser)

ser2 = pd.Series(["Apple_사과", "Banana_바나나", "Cherry_체리", np.nan],
                index=["First ", " Second", " Third", "Fourth"], dtype="string")
print(ser2)
```
