```python
import pandas as pd
```

```python
dict_data = {
    "a": 1,
    "b": 2,
    "c": 3
}

sr = pd.Series(dict_data)

print(type(sr))
print(sr)
```

```python
# 인덱스 배열
## Series객체.index

# 데이터값 배열
## Series객체.values

print(sr.index)
print(sr.values)
```

```python
# 시리즈 인덱스

import pandas as pd

list_data = ["2019-01-02", 3.14, "ABC", 100, True]
sr = pd.Series(list_data)
print(type(sr))
print(sr)
```

```python
idx = sr.index
print(idx)
val = sr.values
print(val)
print(sr.dtype)     # 원소의 자료형
print(len(sr))      # 배열의 크기
print(sr.shape)     # 배열의 형태
print(sr.ndim)      # 배열의 차원
```

```python
import pandas as pd

tup_data = ('영인', '2010-05-01', '여', True)
sr = pd.Series(tup_data, index=['이름', '생년월일', '성별', '학생여부'])

print(sr)
```

```python
# 시리즈 원소 선택
print(sr["이름"])
print(sr[['생년월일', '성별']])
```

```python
print(sr['생년월일':'성별'])
```

```python
# 빈 시리즈
sr = pd.Series()
print(sr)
sr2 = pd.Series(5)
print(sr2)
```
