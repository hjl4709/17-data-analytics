```python
import pandas as pd

dict_data = {
    "c0": [1, 2, 3],
    "c1": [4, 5, 6],
    "c2": [7, 8, 9],
    "c3": [10, 11, 12],
    "c4": [13, 14, 15]
}

df = pd.DataFrame(dict_data)
print(type(df))
print(df)
```

```python
# pandas.DataFrame(2차원배열,
#                  index=행 인덱스 배열,
#                  columns=열 이름 배열
#                  )

arr = ([
    [15, "남", "덕영중"],
    [17, "여", "수리중"]])

df = pd.DataFrame(arr,
                  index=['준서', '예은'],
                  columns=['나이', '성별', '학교'])
print(type(df))
print(df)
```

```python
print(df.index)
print(list(df.index))
print(df.columns)
print(list(df.columns))
```

```python
# 행 인덱스 변경
# DataFrame객체.index = 행인덱스배열
# DataFrame객체.columns = 열인덱스배열

print(df)
```

```python
# 행 인덱스 변경 : DataFrame객체.rename(index={기존인덱스:새 인덱스, ...})
# 열 이름 변경 : DataFrame객체.rename(columns={기존이름:새 이름, ...})

matrix = [[15, '남', '덕영중'], 
          [17, '여', '수리중']]

df = pd.DataFrame(matrix,
                  index=['준서', '예은'],
                  columns=['나이', '성별', '학교'])

df = df.rename(columns={
    '나이':'연령',
    '성별':'남녀',
    '학교':'소속'
    })

df
```

```python
df = df.rename(index={
    '준서':'학생1',
    '예은':'학생2'
    })

df
```
